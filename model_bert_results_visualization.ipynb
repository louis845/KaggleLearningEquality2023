```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Markdown
```

```python
def display_training_data(dframe):
    display(Markdown("**Entropy metrics**"))
    dframe[[x for x in dframe.columns if "entropy" in x]].plot(figsize = (30, 30))
    plt.show()

    display(Markdown("**Training metrics**"))
    dframe[["accuracy", "precision", "recall"]].plot(figsize = (30, 30))
    plt.show()

    display(Markdown("**Test metrics**"))
    dframe[[x for x in dframe.columns if x.startswith("test")]].plot(figsize = (30, 30))
    plt.show()
    
    if len([x for x in dframe.columns if x.startswith("original")])>0:
        display(Markdown("**Original metrics**"))
        dframe[[x for x in dframe.columns if x.startswith("original") and "small" not in x]].plot(figsize = (30, 30))
        plt.show()

        display(Markdown("**Original metrics small**"))
        dframe[[x for x in dframe.columns if x.startswith("original") and "small" in x]].plot(figsize = (30, 30))
        plt.show()

running_data = pd.read_csv("models/direct_model/logfile.csv")
display_training_data(running_data)
```

```python
running_data2 = pd.read_csv("models/overshoot3/logfile.csv")
display_training_data(running_data2)
```

```python
running_data3 = pd.read_csv("models/overshoot2_unbalanced/logfile.csv")
display_training_data(running_data3)
```

```python
running_data4 = pd.read_csv("models/overshoot23_unbalanced/logfile.csv")
display_training_data(running_data4)
```
