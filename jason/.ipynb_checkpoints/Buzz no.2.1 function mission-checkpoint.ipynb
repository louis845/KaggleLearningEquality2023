```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Markdown
```

```python
def data_readlog(filename):
    running_data = pd.read_csv("../models/"+filename+"/logfile.csv", index_col = 0)
```

```python
filename=input("enter file name=")
```

```python
data_readlog(filename)
```

```python
data_len=len(running_data)
display(running_data[["test_precision", "test_recall"]])
```

```python
running_data[["test_precision", "test_recall"]].plot(figsize = (30, 30))

plt.show
```

```python
mini_val=1
for k in range(data_len):
    
    if round(running_data.loc[k].at["test_precision"]-running_data.loc[k].at["test_recall"],5)<mini_val:
        mini_val=abs(running_data.loc[k].at["test_precision"]-running_data.loc[k].at["test_recall"])
        mini_epoch=k
        
print("optimal epoch for mini= " + str(mini_epoch))
print("value of test_precision and test_recall= "+ str(mini_val))
```

```python
running_data[["test_precision", "test_recall"]].plot(figsize = (30, 30))
plt.axvline(x=mini_epoch)
plt.show
```

```python
display((running_data.loc[mini_epoch]))
```

```python
max_val70=0
for k in range(data_len):
    
    if running_data.loc[k].at["test_precision"]<=0.7:
        continue
    
    elif running_data.loc[k].at["test_precision"]>0.7 and running_data.loc[k].at["test_recall"] > max_val70:
    
        max_val70=running_data.loc[k].at["test_recall"]
        max_epoch70=k

print("optimal epoch for test_precision>0.7 = " + str(max_epoch70))
print("value of test_recall= "+ str(max_val70))    
```

```python
display((running_data.loc[max_epoch70]))
```
