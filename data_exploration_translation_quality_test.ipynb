```python
def is_known_useless(x):
    if type(x) == str:
        return (x.startswith("source_id") or (sum(c.isdigit() for c in x) > 7 and sum(c.isalpha() for c in x) > 7)
            or (x.startswith("v") and "." in x))
    return False
```

# Contents bad text

```python
def has_potential_bad_text(x):
    if type(x) != str:
        return False
    return len([tok for tok in x.split() if len(tok) > 15]) > 0

def find_bad_text(x):
    if type(x) != str:
        return "no_bad_text"
    return [tok for tok in x.split() if len(tok) > 15]

lctext = cont.loc[(~cont["text"].isnull()) & cont["text"].apply(has_potential_bad_text) & (contents["language"] != "zh")][["text"]]
btseries = lctext["text"].apply(find_bad_text).rename("bad texts")
btexts = lctext.join(btseries)
```
